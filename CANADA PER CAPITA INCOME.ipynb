### CANADA PER CAPITA INCOME 

#### 1️⃣ Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

#### 2️⃣ Load Dataset

```python
data=pd.read_csv('canada_per_capita_income.csv')
data
```

#### 3️⃣ Prepare Data

```python
x=data[['year']]
y = data['per capita income (US$)']
x
y
```

#### 4️⃣ Train Linear Regression Model

```python
model = LinearRegression()
model.fit(x, y)
```

#### 5️⃣ Make Predictions

```python
y_pred = model.predict(x)
y_pred
```

#### 6️⃣ Model Evaluation

```python
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

print("Coefficient (Slope):", model.coef_[0])
print("Intercept:", model.intercept_)
print("RMSE:", rmse)
print("R² Score:", r2)
```

#### 7️⃣ Visualization

```python
plt.figure(figsize=(8,5))
plt.scatter(x, y)
plt.plot(x, y_pred)
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.title("Canada Per Capita Income - Linear Regression")
plt.show()

```

#### 8️⃣ Predict Future Income (Example: 2025)

```python
future_year = [[2025]]
future_income = model.predict(future_year)
print("Predicted Per Capita Income for 2025:", future_income[0])
```

#### Summary

In this project, Linear Regression was applied to analyze the relationship between year and Canada’s per capita income. The dataset was loaded and explored using Pandas, and the model was built using scikit-learn. Year was considered as the independent variable, while per capita income (US$) was the dependent variable. The model learned the historical trend of income growth and generated predictions. Performance was evaluated using RMSE and R² score, and the results indicated a strong linear relationship. A visualization was created to compare actual income values with the regression line, clearly showing a steady upward trend over time.

#### Conclusion

The Linear Regression model successfully captured the income growth trend in Canada. The high R² score shows that the year is a strong predictor of per capita income. This confirms that income has increased consistently over time. While the model performs well for trend analysis and basic forecasting, real-world economic factors such as inflation, policy changes, and global events are not considered. Therefore, the model is best suited for trend estimation rather than exact future prediction. Overall, Linear Regression proved to be an effective and interpretable method for this economic analysis.
