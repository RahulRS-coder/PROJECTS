### Stock Price Prediction

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
data =pd.read_csv("stock_price_1_year.csv")
data
```

```python
X = data[['Open', 'High', 'Low', 'Volume']]
y = data['Close']
X
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False)
X_train, X_test, y_train, y_test 
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
```

```python
plt.figure(figsize=(10,5))
plt.plot(y_test.values, label="Actual Price")
plt.plot(y_pred, label="Predicted Price")
plt.xlabel("Time")
plt.ylabel("Stock Price")
plt.title("Actual vs Predicted Stock Price")
plt.legend()
plt.show()
```

## Summary

This project focuses on predicting stock closing prices using Linear Regression and historical market data. A one-year dataset containing Open, High, Low, Close, and Volume values was used for analysis. The dataset was preprocessed and split into training and testing sets while preserving the chronological order to maintain the integrity of time-series data.

The Linear Regression model was trained using Open, High, Low, and Volume as input features to predict the closing price. Model performance was evaluated using Mean Absolute Error (MAE) and Root Mean Squared Error (RMSE) metrics. Visualization of actual versus predicted prices helped in understanding how closely the model follows real market trends.

### Conclusion

The results demonstrate that Linear Regression can effectively capture basic relationships in historical stock price data and provide reasonable predictions for closing prices. While the model performs well as a baseline approach, stock prices are influenced by multiple external factors such as market sentiment, economic conditions, and news events, which are not considered in this model.

Therefore, although Linear Regression is useful for understanding trends and building an initial predictive system, more advanced techniques like LSTM, Random Forest, or ARIMA, along with additional technical indicators, can significantly improve prediction accuracy. This project successfully illustrates the application of machine learning in financial data analysis and provides a strong foundation for future enhancements.
